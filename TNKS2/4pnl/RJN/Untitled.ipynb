```python
import os
from IPython.display import display, SVG
os.listdir()
```

```python
from IPython.display import display, HTML
def show_svg(svg1, svg2, text1, text2):
    with open(text1, "r", encoding="utf-8") as file:
        text11 = file.read()  # Считает весь файл в одну строку
    with open(text2, "r", encoding="utf-8") as file:
        text22 = file.read()  # Считает весь файл в одну строку
    
    html_code = f"""
    <table style="width:100%; text-align:center;">
        <tr>
            <td><img src="{svg1}" width="100%" align = center></td>
            <td><img src="{svg2}" width="100%" align = center></td>
        </tr>
        <tr>
            <td style="text-align:left; font-size:10px;"><pre>{text11}</pre></td>
            <td style="text-align:left; font-size:10px;"><pre>{text22}</pre></td>
        </tr>
    </table>
    """
    
    display(HTML(html_code))
```

```python
os.system('obabel -:\"O=C(NCc1ccc(cc1)C2=Nc3ccccc3C(=O)N2)OCc4ccccc4\" -O RJN_SMI.mol2 --gen3D -d')
os.system('obabel RJN_SMI.mol2 -O RJN_SMI.svg -xd -xX -xi')

os.system('obabel RJN.pdb -O RJN_pdb.mol2')
os.system('obabel RJN_pdb.mol2 -O RJN_pdb.svg -xd -xX -xi')
```
