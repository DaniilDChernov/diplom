```python
import MDAnalysis as mda
from MDAnalysis.coordinates.AMCrd import AMCrdWriter
```

```python
# Загрузка траектории GROMACS
u = mda.Universe("topol.tpr", "trajectory.xtc")

# Конвертация траектории в формат AMBER (.mdcrd)
with AMCrdWriter("output.mdcrd", n_atoms=u.atoms.n_atoms) as w:
    for ts in u.trajectory:
        w.write(ts)
```

```python
def check_mdcrd_format(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    incorrect_lines = []
    for index, line in enumerate(lines, start=1):
        numbers = line.split()  # Разделение строки по пробелам
        if len(numbers) < 10:  # Проверяем, меньше ли 10 чисел
            incorrect_lines.append((index, len(numbers), line.strip()))
    
    if incorrect_lines:
        print("Строки с ошибками (меньше 10 чисел):")
        for line_info in incorrect_lines:
            print(f"Строка {line_info[0]}: {line_info[1]} чисел - '{line_info[2]}'")
    else:
        print("Все строки содержат 10 чисел.")

# Укажите путь к вашему файлу mdcrd
file_path = "/home/daniil/Downloads/AMBER/1err_prod.mdcrd"
check_mdcrd_format(file_path)
```

```python
import re

def check_mdcrd_format(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    incorrect_lines = []
    for index, line in enumerate(lines, start=1):
        # Используем регулярное выражение для извлечения чисел
        numbers = re.findall(r'-?\d+\.\d+', line)
        
        if len(numbers) < 10:  # Проверяем, меньше ли 10 чисел
            incorrect_lines.append((index, len(numbers), line.strip()))
    
    if incorrect_lines:
        print("Строки с ошибками (меньше 10 чисел):")
        for line_info in incorrect_lines:
            print(f"Строка {line_info[0]}: {line_info[1]} чисел - '{line_info[2]}'")
    else:
        print("Все строки содержат 10 чисел.")

# Укажите путь к вашему файлу mdcrd
file_path = "/home/daniil/Downloads/AMBER/1err_prod.mdcrd"
check_mdcrd_format(file_path)
```
